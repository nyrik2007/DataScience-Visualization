```python
import matplotlib.pyplot as plt
```

```python
data1=exponential(5,20)
data2=exponential(6,20)
```

```python
plt.plot(data1, label="Первый игорк")
plt.scatter(range(len(data2)),data2, color="red", label="Второй игрок")
plt.title("Результаты", fontdict={"fontsize":20},color="white")
plt.xlabel("Номер попытки", color="white")
plt.ylabel("Результат", color="white")
plt.legend()
plt.xticks(range(0,20,4))
```

# Расширенная визуализация

```python
fig, axes=plt.subplots()
axes.plot(exponential(5,20))
axes.set_title("График")
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3, figsize=(10,5))
for row, row_axes in enumerate(axes):
    for column,ax in enumerate(row_axes):
        ax.plot(exponential(column,20))
        ax.set_title("Холст колонка {}, строка{}".format(column+1, row+1))
fig.tight_layout()  #мы говорим figure что бы все надписи помещались между холстами.
```
